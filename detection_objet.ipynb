2.Charger notre model et faire la prediction 

```python
import os
from ultralytics import YOLO
from IPython.display import Image, display

# Charger le modèle YOLO
model = YOLO('yolov5su.pt')  # Modèle recommandé pour de meilleures performances

# Définir l'image source et le dossier de sortie
source_image = "zidane.jpg"  # Nom de votre image
output_folder = "/content/yolo_results"  # Dossier principal de sauvegarde

# Effectuer la prédiction
results = model.predict(source=source_image, save=True, project=output_folder)

# Chemin du sous-dossier "predict" où les fichiers sont sauvegardés
predict_folder = os.path.join(output_folder, 'predict')

# Vérifier si le sous-dossier existe
if os.path.exists(predict_folder):
    print("Fichiers générés :", os.listdir(predict_folder))

    # Afficher les images générées
    for file in os.listdir(predict_folder):
        if file.endswith(('.jpg', '.png')):  # Filtrer les images
            display(Image(filename=os.path.join(predict_folder, file)))
else:
    print(f"Le dossier {predict_folder} n'existe pas.")
```
